# Arms exports (SIPRI trend indicator values) e Arms imports (SIPRI trend indicator values)

#### Comparação Brazil X Argentina

```python
# EXPORTAÇÃO 

br_export = df[(df['CountryName'] == 'Brazil') & (df['IndicatorName'] == 'Arms exports (SIPRI trend indicator values)')]
ar_export = df[(df['CountryName'] == 'Argentina') & (df['IndicatorName'] == 'Arms exports (SIPRI trend indicator values)')]
data_br_ar_export = pd.DataFrame(columns=['Brazil', 'Argentina'])
br_data_export = br_export.set_index('Year')
ar_data_export = ar_export.set_index('Year')
data_br_ar_export['Brazil'] = br_data_export['Value']
data_br_ar_export['Argentina'] = ar_data_export['Value']

# IMPORTAÇÃO

br_import = df[(df['CountryName'] == 'Brazil') & (df['IndicatorName'] == 'Arms imports (SIPRI trend indicator values)')]
ar_import = df[(df['CountryName'] == 'Argentina') & (df['IndicatorName'] == 'Arms imports (SIPRI trend indicator values)')]
data_br_ar_import = pd.DataFrame(columns=['Brazil', 'Argentina'])
br_data_import = br_import.set_index('Year')
ar_data_import = ar_import.set_index('Year')
data_br_ar_import['Brazil'] = br_data_import['Value']
data_br_ar_import['Argentina'] = ar_data_import['Value']

# GRÁFICO COMPARATIVO

data_br = pd.DataFrame(data={'Brazil Imports': data_br_ar_import['Brazil'], 'Brazil Exports': data_br_ar_export['Brazil']}, index=data_br_ar_import.index)
data_ar = pd.DataFrame(data={'Argentina Imports': data_br_ar_import['Argentina'], 'Argentina Exports': data_br_ar_export['Argentina']}, index=data_br_ar_export.index)
data_br_ar = pd.merge(data_br, data_ar, left_index=True, right_index=True )
data_br_ar.plot(figsize=(13,6), title='Arms Export and Arms Imports: Brazil X Argentina')

```
